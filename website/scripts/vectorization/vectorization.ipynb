
# TF-IDF: A Foundation of Information Retrieval

In order to begin the Psalm retrieval website and model, we need to work on the basics of information retrieval. This will be accomplished using **TF-IDF (Term Frequency - Inverse Document Frequency)**, a statistical measure used to evaluate the importance of a word in a document relative to a collection of documents (the corpus). Once this is finished, we will have vectors that can be used for comparisons later during the information retrieval process.

## Understanding TF-IDF

TF-IDF represents two different calculations that, when combined, result in a score for each word in a given document within a set of documents. The entire calculation is composed of the following:

$$
TF-IDF = Term\ Frequency\ (tf) * Inverse\ Document\ Frequency\ (idf)
$$

### Terminology
Before beginning, we should note some terminology. Here are some of the most common terms and variables used within this project:
- **t** - The targeted word within a document.
- **d** - A document or a set of words.
- **N** - The number of documents in the corpus.
- **corpus** - The collection of all documents.

### TF - Term Frequency
This measures the occurrence of a word within a document. The equation to calculate this is:

$$
TF(t, d) = \frac{\text{Count of term } t \text{ in document } d}{\text{Total number of words in } d}
$$

This measures how often a term \( t \) appears in a document \( d \), normalized by the document length. This can be used for each word in a document and is what can be used to vectorize a document.

## IDF - Inverse Document Frequency
The inverse document frequency represents how informative a word is within the document collection.

### Inverse Document Frequency (IDF)

IDF measures how important a term is across a collection of documents. It is calculated as:

$$
IDF(t) = \log \left( \frac{N}{DF(t)} \right)
$$

#### where:
- **N** is the total number of documents in the collection.
- **DF(t)** is the number of documents that contain the term \( t \).

#### Example:

Suppose we have a collection of 5 documents:

| Document | Content |
|----------|---------|
| D1       | "The cat sat on the mat." |
| D2       | "The dog barked at the cat." |
| D3       | "A cat chased a mouse." |
| D4       | "Dogs and cats are common pets." |
| D5       | "The sun is bright today." |

Now, let's calculate \( IDF \) for the term **"cat"**:
- The term "cat" appears in **4** documents (D1, D2, D3, D4).
- Total number of documents \( N = 5 \).
- Document frequency \( DF(\text{"cat"}) = 4 \).

Using the IDF formula:

$$
IDF(\text{"cat"}) = \log \left( \frac{5}{4} \right) = \log(1.25) \approx 0.096
$$

Since "cat" appears in most documents, its IDF is low, meaning it is not very informative. In contrast, a rarer word (e.g., "sun," which appears in only one document) would have a higher IDF, making it more significant for identifying that document.

## End Product
Once we have these calculations, we can then view any document in the form of a vector as shown in the example below for a single document:

$$
\vec{d_j} = (t_{1,j}, t_{2,j}, ..., t_{n,j})

## Case Sensitivty
Within `NLP`, textual data is sensitive. Each word is split apart from the rest of the document and examined on its own without any context. With this being said, the word `Dog`, and the lowercase version, `dog`, are treated as two totally seperate words. By making everything lowercase, we avoid all of these issues. We need to do with for our data. 


````python
# turning all letters lowercase
psalms['cleaned_verse'] = psalms['cleaned_verse'].str.lower()

psalms.head(5)
````


## Stop Words
These are words that occur in almost every document. They are not words that can be used to determine the meaningg of a document. Some of these words include, and not limited to **is**, **and**, and **the**. Within the book [Natural Language Processing with Python](https://www.nltk.org/book/), we can see see that within the `NLTK` python package, we can import a corpus of defined stop words and pull these words out of the **cleaned_verse** column of our dataframe. 


````python
from nltk.corpus import stopwords
import re

# Downloading the corpus of stop words
nltk.download('stopwords')

# Load English stopwords
stop_words = set(stopwords.words("english"))

# Extracting all stop words from "clean_verse" colunms
for index, row in psalms.iterrows():
    verse = row['cleaned_verse']
    
    # Split the verse into words
    words = verse.split()
    
    # Initialize the new verse without stopwords
    new_verse = ""
    
    # Loop over words and add to new_verse if not a stopword
    for word in words:
        if word.lower() not in stop_words:  # .lower() ensures case-insensitive matching
            new_verse += " " + word
        # Update the row with the new cleaned verse
    psalms.at[index, 'cleaned_verse'] = new_verse.strip()  # strip() removes leading/trailing spaces

psalms.head(5)
````

````python
# Symbols to remove (punctuation, etc.)
symbols = "!\"#$%&()*+-.,/:;<=>?@[\]^_`{|}~\n"
for symbol in symbols:
    psalms["cleaned_verse"] =  psalms["cleaned_verse"].replace(symbol, ' ')
    
psalms
````

````python
for index, row in psalms.iterrows():
    verse = row['cleaned_verse']
    
    # Split the verse into words
    words = verse.split()
    
    # Initialize the new verse without stopwords
    new_verse = ""
    
    # Loop over words and add to new_verse if not a stopword
    for word in words:
        if len(word) > 1:
            new_verse += " " + word
        # Update the row with the new cleaned verse
    psalms.at[index, 'cleaned_verse'] = new_verse.strip()  # strip() removes leading/trailing spaces

psalms.head(10)
````

## Stemming

Stemming the process of transforming all words into their root. This ensures that similar words are treated the same. As an example, **cooking something** and **cooked something** are pretty much the same thing in terms of the action to cook something. Without any cleaning or processing, these words would be treated as three separate things: `cook`, `cooked`, and `cooking`. This could prevent a document from being retrieved if the query does not contain the exact form of the word `cook`.

By using stemming, we convert different variations of a word into its root form, improving search accuracy and consistency in text analysis. There are a couple of different stemmers we can use, and will be outlined below. 
### Porter-Stemmer
This stemmer targets and identifies affixes and suffixes of words and removes them. From the NLP text book, referenced above, it says this is a good stemmer for searches because of the indexing it does and the support it gives for searches, using alternative forms of a word.

### Lancaster-Stemmer
The Lancaster Stemmer works to reduce words to their root form. It is more aggressive than other stemmers, such as Porter Stemmer, meaning it can sometimes over-stem words, reducing them to very short roots.

From the textbook [Natural Language Processing with Python](https://www.nltk.org/book/), we are given examples of both stemmers. After reading about both in the text, we are going to use the `Porter-Stemmer` for our data. This will be best for information retrieval.

## Lemmatization
This is a similar method to reduce words to their root word. The difference is found in the output. Within lemmatization, the word is only reduced to its root word if and only the root word exists. This still is a very useful mitigation to treat all variations of a single word as the same thing. Below is a table to illustrate the difference between **Stemming** and **Lemmatization**.

 For this project, we are going to lemmatize the text, as it preserves the text better. Below is a mix of my own code and code from the text book that is being referenced constantly. 


````python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet

# Download necessary resources
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')

# %%
# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Function to lemmatize words in a given text
def lemmatize_text(text, lemmatizer, pos=wordnet.NOUN):
    words = word_tokenize(text)  # Tokenize the text into words
    lemmatized_words = [lemmatizer.lemmatize(word, pos=pos) for word in words]  # Lemmatize each word
    return " ".join(lemmatized_words)
````

````python
# Apply lemmatization to each verse, passing the lemmatizer explicitly
psalms["cleaned_verse"] = psalms["cleaned_verse"].apply(lemmatize_text, lemmatizer=lemmatizer)

# Check the first 10 rows of the dataframe
psalms.head(10)
````

## Converting Numbers
The last step in preparing the text for vectorization is converting any numbers to their word form. If we skip this step, our model may treat a query or text like **"50 dollars"** as two separate entities: *"50"* and *"dollars"*, rather than recognizing them as a single concept. There should not be too many numbers within the psalms, but we are going to implement this anyways, and the *Medium* article does. 


````python
#!pip install num2words
````

````python
from num2words import num2words
import re

def convert_numbers_to_words(text):
    """Convert all numbers in a given text to their word form using num2words."""
    if not isinstance(text, str):
        return text  # Return as-is if not a string

    # Replace each number with its word form
    return re.sub(r'\d+', lambda x: num2words(int(x.group())), text)


# Apply function to the psalms dataframe
psalms["cleaned_verse"] = psalms["cleaned_verse"].apply(convert_numbers_to_words)

psalms.head(10)
````
